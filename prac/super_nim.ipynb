```python
import random


board = [str(digit) for digit in range(1, 65)]
board_with_chips = []
                                                   

def create_board_with_chips():
    
    chips_quantity = random.randint(5, 32)
    
    for _ in range(chips_quantity):
        r = random.randint(0, 63)
        board[r - 1] = 'Ф'
        
    for el in range(len(board)):
        if len(str(board[el])) < 2:
            board[el] = ' ' + str(board[el])
        
    for el in range(len(board)):
        board_with_chips.append(board[el])
        
    
def drew_board_with_chips():
    
    print('-' * 41)
    for i in range(8):       
        print('|', board_with_chips[0 + i * 8], '|', board_with_chips[1 + i * 8], '|', board_with_chips[2 + i * 8], '|', board_with_chips[3 + i * 8], '|', board_with_chips[4 + i * 8], '|', board_with_chips[5 + i * 8], '|', board_with_chips[6 + i * 8], '|', board_with_chips[7 + i * 8], '|')
        print('-' * 41)
                
        
def check_win():
    
    if 'Ф' in board_with_chips or ' Ф' in board_with_chips:
        return False
        
    else:
        return True
    
    
        
def move(num):
    
    line = []
    column = []
    
    if 'l' in num:
        num = int(num[1:])
        line = board_with_chips[(num - 1):(num + 7)]
        #print(line)
        for el in range(len(line)):
            if line[el] == 'Ф' or line[el] == ' Ф':
                board_with_chips[el + (num - 1)] = '  '
            else:
                continue
        
    elif 'c' in num:
        num = int(num[1:])
        column = board_with_chips[(num - 1)::8]
        #print(column)
        for el in range(len(column)):
            if column[el] == 'Ф' or column[el] == ' Ф':
                board_with_chips[(el * 8) + (num - 1)] = '  '
            else:
                continue
        
    else:
        print('Wrong input')
        return False

        
def main():
    
    step = 0
    create_board_with_chips()
    
    if step % 2 != 0:
        player = 'player1'
    elif step % 2 == 0:
        player = 'player2'
    
    while True:
        
        step += 1 
        drew_board_with_chips()
        
        winner = check_win()
        if winner:
            print(f'{player} won!')
            break
                
        print('Example of a move: for a line - l1, l9, l25... for a column - c1, c2, c6')
        num = input('Make your move: ')
        move(num)
        #print(board_with_chips)
        
main()       
```
