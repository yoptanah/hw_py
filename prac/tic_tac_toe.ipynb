```python
import copy

board = [1, 2, 3, 4, 5, 6, 7, 8, 9]

board_og = copy.deepcopy(board)

win_combinations = [(1, 2, 3), (4, 5, 6), (7, 8, 9), (1, 5, 9), (3, 5, 7), (1, 4, 7), (2, 5, 8), (3, 6, 9)]

def draw_board():
    print('-------------')
    for i in range(3):
        print('|', board[0 + i * 3], '|', board[1 + i * 3], '|', board[2 + i * 3], '|')
        print('-------------')

def player_input(token):
    while True:
        value = input('Вместо какого числа подставить '+ token + ': ')
        if not (value in '123456789'):
            print('Неправильный ввод, повторите')
            continue
        value = int(value)
        if value > 9:
            print('Неправильный ввод, повторите')
            continue
        if str(board[value - 1]) in 'XO':
            print('Клетка уже занята')
            continue
        board[value - 1] = token
        break

def check_win():
    for cell in win_combinations:
        if (board[cell[0]-1]) == (board[cell[1]-1]) == (board[cell[2]-1]):
            return board[cell[0] - 1]
    else:
        return False

def main():
    step = 0
    while True:
        draw_board()
        if step % 2 == 0:
            player_input('X')
        else:
            player_input('O')
        if step > 3:
            winner = check_win()
            if winner:
                draw_board()
                print(f'{winner} выиграл')
                break
        step += 1
        if step > 8:
            draw_board()
            print('Ничья')
            break
            
main()
```
